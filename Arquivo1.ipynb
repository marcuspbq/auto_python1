# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# Passo a passo para resolver o desafio:
import pyautogui
import pyperclip
import time

# tempo/delay
pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema da empresa(link do drive)

pyautogui.hotkey("ctrl","t") # abrir uma aba no navegador
pyperclip.copy("https://1drv.ms/u/s!AkepAYh_UbLIkYpIyT_BnajDugnMow?e=oVnxEA") # copia com caracteres especiais
pyautogui.hotkey("ctrl","v") # cloa o link
pyautogui.press("enter") # aperta enter

# Carregar o sistema
time.sleep(10)

# Passo 2: Navegar no sistema (até a pasta exportar)
pyautogui.click(x=528, y=366) # Click para abrir a planilha
time.sleep(2)

# Passo 3: Fazer o download da base de vendas
pyautogui.click(x=421, y=303) # Selecionar a planilha
pyautogui.click(x=506, y=170) # Clicar no botão download
time.sleep(5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Importar a base de vendas pro Python
import pandas as pd
tabela = pd.read_excel(r"C:\Users\marcu\Downloads\Vendas - Dez.xlsx") # O "r" é para que o python não tente ler caracteres especiais como "\", porque o python tem vários caracteres especiais que tem funçãoes específicas
display(tabela) # display é uma biblioteca do Júpyter que substitui o print por exibir texto formatado

# Passo 5: Calcular o faturamento e quantidade de produtos vendidos
faturamento = tabela["Valor Final"].sum() #.count (contar) .mean (média) .max(máximo) .min(mínimo)
qtde_produtos = tabela["Quantidade"].sum() #.count (contar) .mean (média) .max(máximo) .min(mínimo)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 6: Enviar email para a diretoria

# Abrir email
pyautogui.hotkey("ctrl","t") # abrir uma aba no navegador
pyperclip.copy("https://outlook.live.com/mail/0/") # copia com caracteres especiais
pyautogui.hotkey("ctrl","v") # coloa o link
pyautogui.press("enter") # aperta enter
time.sleep(5)
pyautogui.click(x=127, y=171) # Clicar no botão escrever
time.sleep(2)
pyperclip.copy("ti.marc@outlook.com") # copia o endereço de email com caracteres especiais
pyautogui.hotkey("ctrl","v") # coloa o endereço de email
pyautogui.press("tab") # Alternar no campo assunto
pyperclip.copy("Relatório de vendas") # copia o assunto com caracteres especiais
pyautogui.hotkey("ctrl","v") # colo o assunto do email
pyautogui.press("tab") # Alternar para o corpo do email
# Escrever o email
texto = f"""
Prezados, bom dia!

O faturamento de ontem foi de R$ {faturamento:,.2f}
A quantidade foi de {qtde_produtos:,}

Abs
Marcus Paulo
"""# Para colocar informações dinâmicas , coloca o "f" no início e as variáveis entre chaves
pyperclip.copy(texto) # copia a variável com caracteres especiais
pyautogui.hotkey("ctrl","v") # coloa o texto

pyautogui.hotkey("ctrl","enter") # Enviar email
```
